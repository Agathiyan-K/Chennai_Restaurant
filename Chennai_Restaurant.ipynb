```python
import gdown

# Google Drive file ID from your link
file_id = "1uZt10B_f16oTj_u_EM-VCUhI6nmGQzuG"
url = f"https://drive.google.com/uc?id={file_id}"

# Download the file
gdown.download(url, "zomato_chennai.csv", quiet=False)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("zomato_chennai.csv")

# View basic info
print(df.shape)
print(df.columns)
df.head()
```

```python
df.columns = df.columns.str.strip()  # Clean column names
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
import ast

def safe_eval(val):
    if isinstance(val, str):
        try:
            return ast.literal_eval(val)
        except:
            return [val]  # fallback if it's not a list-like string
    return val

# Apply safely
for col in ['Cuisine', 'Top Dishes', 'Features']:
    df[col] = df[col].apply(safe_eval)
```

```python
# Explode the columns
df_exploded = df.explode('Cuisine')
df_exploded = df_exploded.explode('Top Dishes')
df_exploded = df_exploded.explode('Features')

# Optional: clean up extra whitespace
for col in ['Cuisine', 'Top Dishes', 'Features']:
    df_exploded[col] = df_exploded[col].astype(str).str.strip()
```

```python
import pandas as pd
from dash import Dash, dcc, html, Input, Output
import dash_bootstrap_components as dbc
import plotly.express as px
import plotly.graph_objects as go
from dash import dash_table

# Load your exploded dataset
df = df_exploded  # Assuming df_exploded is already defined

# Initialize Dash app
app = Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

# App layout
app.layout = dbc.Container([
    html.H2("Zomato Chennai Restaurant Explorer", className="text-center my-3"),

    dbc.Row([
        dbc.Col([
            html.Label("Select Location"),
            dcc.Dropdown(
                id='location-dropdown',
                options=[{'label': loc, 'value': loc} for loc in sorted(df['Location'].unique())],
                placeholder="Choose a location"
            )
        ])
    ], className="mb-3"),

    dbc.Row([
        dbc.Col([
            html.Label("Select Cuisine"),
            dcc.Dropdown(id='cuisine-dropdown', placeholder="Choose a cuisine")
        ])
    ], className="mb-3"),

    dbc.Row([
        dbc.Col([
            html.Label("Select Restaurant(s)"),
            dcc.Dropdown(id='restaurant-dropdown', multi=True, placeholder="Choose restaurant(s)")
        ])
    ], className="mb-3"),

    dbc.Row([
        dbc.Col([
            html.Div(id='restaurant-details')
        ])
    ]),

    # Visualizations
    dbc.Row([
        dbc.Col([
            dcc.Graph(id='price-comparison-chart')
        ], width=4),
        dbc.Col([
            dcc.Graph(id='rating-comparison-chart')
        ], width=4),
        dbc.Col([
            dcc.Graph(id='heatmap-chart')
        ], width=4)
    ])
], fluid=True)

# Callback to update cuisine options
@app.callback(
    Output('cuisine-dropdown', 'options'),
    Input('location-dropdown', 'value')
)
def update_cuisine_dropdown(selected_location):
    if selected_location:
        filtered = df[df['Location'] == selected_location]
        cuisines = set()
        for items in filtered['Cuisine']:
            if isinstance(items, str):
                for i in items.strip("[]").replace("'", "").split(","):
                    cuisines.add(i.strip())
        return [{'label': c, 'value': c} for c in sorted(cuisines)]
    return []

# Callback to update restaurant options
@app.callback(
    Output('restaurant-dropdown', 'options'),
    [Input('location-dropdown', 'value'),
     Input('cuisine-dropdown', 'value')]
)
def update_restaurant_dropdown(loc, cuisine):
    if loc and cuisine:
        filtered = df[df['Location'] == loc]
        filtered = filtered[filtered['Cuisine'].str.contains(cuisine, na=False)]
        restaurants = sorted(filtered['Name of Restaurant'].unique())
        return [{'label': name, 'value': name} for name in restaurants]
    return []

# Callback to display data and visuals
@app.callback(
    [Output('restaurant-details', 'children'),
     Output('price-comparison-chart', 'figure'),
     Output('rating-comparison-chart', 'figure'),
     Output('heatmap-chart', 'figure')],
    Input('restaurant-dropdown', 'value')
)
def display_restaurant_details(selected_restaurants):
    if selected_restaurants:
        filtered_df = df[df['Name of Restaurant'].isin(selected_restaurants)]

        grouped_df = filtered_df.groupby('Name of Restaurant').agg({
            'Address': 'first',
            'Top Dishes': lambda x: ', '.join(x.dropna().unique()),
            'Features': lambda x: ', '.join(x.dropna().unique()),
            'Price for 2': 'first',
            'Dining Rating': 'mean',
            'Dining Rating Count': 'sum',
            'Delivery Rating': 'mean',
            'Delivery Rating Count': 'sum'
        }).reset_index()

        # Table
        table = dash_table.DataTable(
            id='restaurant-comparison-table',
            columns=[{'name': col, 'id': col} for col in grouped_df.columns],
            data=grouped_df.to_dict('records'),
            style_table={'height': '400px', 'overflowY': 'auto'},
            style_cell={'textAlign': 'left', 'padding': '8px'},
            style_header={'backgroundColor': 'lightgrey', 'fontWeight': 'bold'}
        )

        # Price Bar Chart
        price_comparison_fig = px.bar(
            grouped_df,
            x='Name of Restaurant',
            y='Price for 2',
            title='Price for 2 Comparison',
            labels={'Price for 2': 'Price for 2 (₹)'}
        )

        # Ratings Bar Chart
        rating_comparison_fig = go.Figure()
        rating_comparison_fig.add_trace(go.Bar(
            x=grouped_df['Name of Restaurant'],
            y=grouped_df['Dining Rating'],
            name='Dining Rating',
            marker_color='blue'
        ))
        rating_comparison_fig.add_trace(go.Bar(
            x=grouped_df['Name of Restaurant'],
            y=grouped_df['Delivery Rating'],
            name='Delivery Rating',
            marker_color='orange'
        ))
        rating_comparison_fig.update_layout(
            title='Dining and Delivery Rating Comparison',
            barmode='group',
            xaxis_title='Restaurant',
            yaxis_title='Rating'
        )

        # Plotly Heatmap (correlation)
        corr_df = grouped_df[['Price for 2', 'Dining Rating', 'Delivery Rating']]
        corr_matrix = corr_df.corr().round(2)

        heatmap_fig = go.Figure(data=go.Heatmap(
            z=corr_matrix.values,
            x=corr_matrix.columns,
            y=corr_matrix.index,
            colorscale='RdBu',
            zmin=-1,
            zmax=1,
            text=corr_matrix.values,
            texttemplate="%{text}",
            colorbar=dict(title="Correlation")
        ))
        heatmap_fig.update_layout(
            title='Correlation Heatmap: Price, Dining, and Delivery Ratings',
            xaxis_title='',
            yaxis_title='',
            height=400
        )

        return table, price_comparison_fig, rating_comparison_fig, heatmap_fig

    return html.P("Select at least one restaurant to see details."), {}, {}, go.Figure()

# Run app
if __name__ == '__main__':
    app.run(debug=True)
```
